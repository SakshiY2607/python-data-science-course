- kde and hist are almost similar
- a simple visualisation


```python
weather['Precipitation'].plot(figsize=(10,5),kind='kde') 
```

```python
weather['MaxTemp'].plot(figsize=(10,5),kind='kde')  #(bins willgive number of peices) 
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,6),kind='hist',color=['green','hotpink'],alpha=.5,bins=25) #alpha gives transparency
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,8),kind='hist',color=['green','hotpink'],bins=25,alpha=.5,subplots=True,Layout=(2,1)) #alpha gives transparency
```

```python
weather[['MaxTemp','MinTemp']].plot(figsize=(20,8),kind='hist',color=['green','hotpink'],bins=25,alpha=.5,subplots=True,Layout=(1,2))
```
