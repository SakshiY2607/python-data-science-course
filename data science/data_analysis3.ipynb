# Group By 
- pivot table are upgraded version of group by

```python
#per occupation average experience
df.groupby('OCCUPATION')['EXPERIENCE'].mean().plot(kind='pie') 
```

```python
df.groupby('OCCUPATION')['EXPERIENCE','WAGE'].mean().plot()
```

```python
df.groupby('OCCUPATION')['EXPERIENCE'].agg(['mean','min','max','median','std']) #in agg we pass a list
```

```python
df.groupby('OCCUPATION')['EXPERIENCE','WAGE'].agg(['mean','min','max','median','std']).plot(kind='box')
```
